```python
import cv2
img_name = 'test.jpg'

img = cv2.imread(img_name)

d = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

n_boxes = len(d['level'])

for i in range(n_boxes):

    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])

    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

 

plt.imshow(img)
```
