```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
import ipywidgets as widgets
# sin波のデータ生成
X = np.linspace(0, 10, 1000).reshape(-1, 1)
y = np.sin(X).ravel() #sin波の一次元データに変換
# スライダーウィジェットの作成
test_size_slider = widgets.FloatSlider(value=0.2, min=0.1, max=0.9, step=0.1, description='Test Size:')
def update_plot(test_size):
    # 訓練データとテストデータに分割
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    model = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=10000, random_state=42)
    model.fit(X_train, y_train)

    # 予測
    y_pred = model.predict(X_test)

    # 結果のプロット
    plt.figure(figsize=(10, 6))
    plt.scatter(X_test, y_test, color='red', label='True Values')
    plt.scatter(X_test, y_pred, color='blue', label='Predicted Values')
    plt.legend()
    plt.xlabel('X')
    plt.ylabel('sin(X)')
    plt.title(f'Sin Wave Prediction using MLPRegressor (Test Size: {test_size})')
    plt.show()
# スライダーの値が変わるたびにプロットを更新
widgets.interact(update_plot, test_size=test_size_slider)
```

```python
# cos波のデータ生成
X_cos = np.linspace(0, 10, 1000).reshape(-1, 1)
y_cos = np.cos(X_cos).ravel()
y_cos_pred = model.predict(X_cos)
# 結果のプロット
plt.figure(figsize=(10, 6))
plt.scatter(X_cos, y_cos, color='green', label='True Cos Values')
plt.scatter(X_cos, y_cos_pred, color='purple', label='Predicted Cos Values')
plt.legend()
plt.xlabel('X')
plt.ylabel('cos(X)')
plt.title('Cos Wave Prediction using MLPRegressor')
plt.show()
```
