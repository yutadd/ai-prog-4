```python
from typing import List
def moji_bigram(orig:str)->List[str]:
    return [orig[i]+orig[i+1] for i in range(len(orig)-1)]
def tango_bigram(orig:str)->List[str]:
    tangos=orig.split(" ")
    return [tangos[i]+tangos[i+1] for i in range(len(tangos)-1)]
orig="paraparaparadise"
orig2="paragraph"
mojib=set(moji_bigram(orig))
mojib2=set(moji_bigram(orig2))
print("X : ",mojib)
print("Y : ",mojib2)
print("和集合:", mojib | mojib2)
print("積集合:",  mojib & mojib2)
print("差集合:", mojib - mojib2)
```

```python
import re
def cipher(orig:str)->str:
    #TODO: 正規表現使わずに数値の範囲を使うようにすると効率いいのは知ってる。
    return "".join([chr(219-ord(c)) if re.match("[a-zA-Z]", c) else c for c in orig])
print(cipher("goes to the ocean."))
```

```python

import random
def randomize_only_inside_of_str(orig:str)->str:
    start=orig[0]
    end=orig[-1]
    middle = list(orig[1:-1])
    random.shuffle(middle)
    return start + ''.join(middle) + end
print(randomize_only_inside_of_str("I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."))
```

```python
orig_filename=r".\.datasets\popular-names.txt"
def read_popular_names()->str:
    with open(orig_filename,"r") as f:
        content=f.read()
        f.close()
        return content;
def save_file_as_another_name(content:str,filename:str):
    with open(filename,"w") as f:
        f.write(content)
        f.close()
def count_wr(text:str)->int:
    return text.count("\n")
count_wr(read_popular_names())
```

```python
from collections import Counter

orig = read_popular_names().split("\n")
first_column = [r.split("\t")[0] for r in orig if r]
frequency = Counter(first_column)
sorted_frequency = sorted(frequency.items(), key=lambda x: x[1], reverse=True)
for item in sorted_frequency:
    print(f"{item[0]}: {item[1]}")
```

```python
import json
orig_filename=r".\.datasets\jawiki-country.json"
def read_jawiki_file()->dict:
    with open(orig_filename,"r",encoding="utf-8") as f:
        content=f.read()
        f.close()
        try:
            json_content = json.loads(content)
            result = {item["title"]: item["text"] for item in json_content}
            return result
        except json.JSONDecodeError as e:
            return {"error": f"JSONDecodeError: {e}"}
print(read_jawiki_file()["エジプト"])
```
